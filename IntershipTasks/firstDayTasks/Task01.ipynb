```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("vikrishnan/iris-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd

dataFarme = pd.read_csv('iris.csv')

dataFarme
```

```python
rows, columns = dataFarme.shape
print(f"The dataset contains {rows} rows and {columns} columns.")
```

```python
columnNames = dataFarme.columns
print("The dataset contains the following columns:")
for col in columnNames:
    print(f"  -{col}")
```

```python
dataFarme.head()
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(13, 5))
plt.suptitle("THE ANALYSIS OF RELATIONSHIP BETWEEN VARIABLES", fontsize=16, fontweight='bold')

# First scatter plot: SepalLength vs SepalWidth
x1 = dataFarme['SepalLengthCm']
y1 = dataFarme['SepalWidthCm']
plt.subplot(1, 2, 1)
plt.scatter(x1, y1, color='blue')
plt.title("Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid()

# Second scatter plot: PetalLength vs PetalWidth
x2 = dataFarme['PetalLengthCm']
y2 = dataFarme['PetalWidthCm']
plt.subplot(1, 2, 2)
plt.scatter(x2, y2, color='green')
plt.title("Petal Length vs Width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid()

# Show plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.suptitle("HISTOGRAM PLOT FOR EXAMINING DATA DISTRIBUTION", fontsize=16, fontweight='bold')

plt.subplot(2, 2, 1)
plt.hist(x1, color='blue', edgecolor='black')
plt.xlabel("Sepal Length (cm)", fontsize=9)
plt.ylabel("Frequency", fontsize=9)
plt.title("Sepal Length Distribution")
plt.grid()

plt.subplot(2, 2, 2)
plt.hist(x2, color='lightgreen', edgecolor='black')
plt.xlabel("Petal Length (cm)", fontsize=9)
plt.ylabel("Frequency", fontsize=9)
plt.title("Petal Length Distribution")
plt.grid()

plt.subplot(2, 2, 3)
plt.hist(y1, color='yellow', edgecolor='black')
plt.xlabel("Sepal Width (cm)", fontsize=9)
plt.ylabel("Frequency", fontsize=9)
plt.title("Sepal Width Distribution")
plt.grid()

plt.subplot(2, 2, 4)
plt.hist(y2, color='pink', edgecolor='black')
plt.xlabel("Petal Width (cm)", fontsize=9)
plt.ylabel("Frequency", fontsize=9)
plt.title("Petal Width Distribution")
plt.grid()

plt.tight_layout()
plt.show()
```

```python
data = [x1, x2, y1, y2]
labels = ['Sepal Length', 'Petal Length', 'Sepal Width', 'Petal Width']

# Create boxplot
plt.figure(figsize=(12, 6))
plt.boxplot(data, patch_artist=True, 
            boxprops=dict(facecolor='#0800F0', color='black', linewidth=2),
            medianprops=dict(color='white', linewidth=2),
            whiskerprops=dict(color='black', linewidth=2),
            capprops=dict(color='black', linewidth=2),
            flierprops=dict(markerfacecolor='red', marker='o', markersize=5))

# Add title and labels
plt.title("BOXPLOT TO DETECT OUTLIERS & SPREAD OF VALUES.", fontsize=16, fontweight='bold')
plt.xticks([1, 2, 3, 4], labels, fontsize=12)
plt.ylabel("Centimeters", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
